```python
name = input("Enter your name: ")
score = float(input("Enter your test score (0-100): "))

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Hi {name}! You scored a {score}%. That is a {grade}.")
```
